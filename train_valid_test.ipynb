```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import random as ra
```

```python
train_data = pd.read_csv("gwas_data_set=1.csv", sep=';')
train_data
```

```python
train_X=train_data.iloc[:,[14,16,18,20,22,24]]
train_X
# train_X=train_data.iloc[:,[24]]
# train_X
```

```python
lenn=len(train_data)
train_Y=[]
for i in range(lenn):
    if(train_data.iloc[i][0]=='n'):
        train_Y.append(1)
    else: train_Y.append(0)
    
train_Y=np.array(train_Y)
print(train_Y)
        
```

```python
valid_data = pd.read_csv("gwas_data_set=2.csv", sep=';')
valid_data
```

```python
valid_X=valid_data.iloc[:,[14,16,18,20,22,24]]
valid_X
# valid_X=valid_data.iloc[:,[24]]
# valid_X
```

```python
lenn=len(valid_data)
valid_Y=[]
for i in range(lenn):
    if(valid_data.iloc[i][0]=='n'):
        valid_Y.append(1)
    else: valid_Y.append(0)
    
valid_Y=np.array(valid_Y)
print(valid_Y)
```

```python
test_data = pd.read_csv("gwas_data_set=3.csv", sep=';')
test_data
```

```python
test_X=test_data.iloc[:,[14,16,18,20,22,24]]
test_X
# test_X=test_data.iloc[:,[24]]
# test_X
```

```python
lenn=len(test_data)
test_Y=[]
for i in range(lenn):
    if(test_data.iloc[i][0]=='n'):
        test_Y.append(1)
    else: test_Y.append(0)
    
test_Y=np.array(test_Y)
print(test_Y)
```

```python
import time
from keras.layers import Dropout
```

```python
start=time.time()

model = Sequential()
model.add(Dense(32, input_dim=6, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))

model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(train_X, train_Y, validation_data=(valid_X,valid_Y), epochs=100, batch_size=32)
# evaluate the keras model
_, accuracy = model.evaluate(train_X, train_Y, verbose=0)
print('Accuracy: %.2f' % (accuracy*100))

end=time.time()
print(end-start)
```

```python
_, accuracy = model.evaluate(test_X, test_Y, verbose=0)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
from keras.models import model_from_yaml
```

```python
model_yaml = model.to_yaml()
with open("new_train_valid_model2.yaml", "w") as yaml_file:
    yaml_file.write(model_yaml)
model.save_weights("new_train_valid_model2.h5")
print("Saved model to disk")
```

```python
yaml_file = open('new_train_valid_model2.yaml', 'r')
loaded_model_yaml = yaml_file.read()
yaml_file.close()
loaded_model = model_from_yaml(loaded_model_yaml)
loaded_model.load_weights("new_train_valid_model2.h5")
print("Loaded model from disk")
loaded_model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
score = loaded_model.evaluate(train_X, train_Y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1] * 100))
score = loaded_model.evaluate(test_X, test_Y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1] * 100))
```
